# **1. What Is Prompt Engineering?**
- Prompt engineering is the practice of designing clear, structured inputs that guide AI systems to produce accurate, useful, and predictable outputs.  
In short: **better instructions → better results.**  
It combines communication skills, critical thinking, and an understanding of how AI models interpret language.

---

# **2. Why It Matters**
- AI systems respond differently depending on phrasing and context.  
- Strong prompting improves efficiency, creativity, and correctness.  
- Good prompts help you control tone, format, and depth of the output.  
- It’s a foundational skill for anyone collaborating with AI—no coding required.

---

# **3. Core Skills for Effective Prompting**
- **Clarity:** Use precise language and define what you want.  
- **Context Awareness:** Provide relevant background or constraints.  
- **Iteration:** Refine prompts based on results rather than expecting perfection on the first try.  
- **Decomposition:** Break complex tasks into smaller, manageable steps.  
- **Role Framing:** Assign the AI a role (e.g., “act as a tutor,” “act as a researcher”) to shape the response.  
- **Output Structuring:** Specify formats (lists, tables, sections) for faster, cleaner results.

# **4. Best Practices**
- **Be specific, not vague:** Replace “Write about climate change” with “Give a 3-point summary of climate change impacts on coastal cities.”  
- **Set constraints:** Time period, tone, audience, or length.  
- **Request reasoning or steps when appropriate:** Helps ensure accuracy and transparency.  
- **Use examples:** Provide a sample answer for style or structure.  
- **Iterate:** Ask the AI to revise, expand, shorten, simplify, or change tone.  
- **Check assumptions:** Ask the model to identify missing info or ambiguities.
